1.1 simple linear regression models the relationship between a single predictor and outcome variable by fitting a straight line to the data. multiple linear regression includes multiple independant variables to predict the outcome. the benefits of multiple linear regression is that it can considers multiple factors which could show more complex relationships between the variables, this is helpful when the outcome is possible affected by multiple variables which allows a better understanding between relationship and influencing factors

1.2 the difference between using a continuous variable and an indicator variable is that a continuous variable shows the change in outcome per each increase in unit in the x axis. it is in the form

y = beta0 + beta1x + epsilon 

where y is the outcome, beta0 is the intercept, beta1 is the slope, x is the predictor variable and epsilon is the error term. an indicator variable is used for categorical data with 2 levels and for each catagory the indicator variable will either be 0 or 1. it is in the form

y = beta0 + beta1.1(predictor) + epsilon

where everything is the same except 1(predictor) is an indicator variable that is either 1 if the conditions are met, or 0 otherwise

1.3 adding both an indicator variable alongside a continuous variable in multiple linear regression changes the model by accounting for a continuous trend and a shift with 2 categories with seperate intercepts. simple linear regression:

y = beta0 + beta1x + epsilon

multiple linear regression

y = beta0 + beta1x + beta2.1(predictor) + epsilon

here beta0 is the intercept when the predictor = 0 and and beta0 + beta2 is the intercept when predictor = 1. so there are 2 intercepts using the same trend line

1.4 adding an interaction term between a continuous and an indicator variable in multiple linear regression allows the model to fit different slopes for each category rather than just different intercepts. mutiple linear regression with an interaction term:

y = beta0 + beta1x + beta2.1(predictor) + beta3.1(predictor)x

here when predictor = 0 then there is a linear regression for the first category with beta0 as the intercept and beta1 as the slope. when predictor = 1 then the linear regression is for the second category with beta0 + beta2 as the intercept and (beta1 + beta3) as the slope.

1.5 when a multiple linear regression model uses non-binary categorical data, it must add additional indicator variables to represent each catagory (number of catagories - 1 = number of indicator variables) . if there are 3 catagories (cat) then 2 indicator variables (ind)

for cat1, ind1=1, ind2=0
for cat2, ind1=0, ind2=1
for cat3, ind1=0, ind2=0

```python
import numpy as np
from sklearn.model_selection import train_test_split

fifty_fifty_split_size = int(pokeaman.shape[0]*0.5)

# Replace "NaN" (in the "Type 2" column with "None")
pokeaman.fillna('None', inplace=True)

np.random.seed(130)
pokeaman_train,pokeaman_test = \
  train_test_split(pokeaman, train_size=fifty_fifty_split_size)
pokeaman_train
```

this cell sets up the data, it shows the data table and the number of rows and columns, replaces the missing values, it also splits the data set into fifty-fifty (training and testing) and also sets a seed such that the split is reproducible.

```python
model_spec3 = smf.ols(formula='HP ~ Attack + Defense', 
                      data=pokeaman_train)
model3_fit = model_spec3.fit()
model3_fit.summary()
```

this cell fits a simple linear regression model that predicts HP based on Attack + Defense and shows the fit.summary().

```python
yhat_model3 = model3_fit.predict(pokeaman_test)
y = pokeaman_test.HP
print("'In sample' R-squared:    ", model3_fit.rsquared)
print("'Out of sample' R-squared:", np.corrcoef(y,yhat_model3)[0,1]**2)
```

this cell calculates the in-sample and out-of-sample values, the in-sample r-squared value is how well the model explains the variability in HP on the training data
the out-of-sample r-squared value indicates how well the model fits using new data

```python
model4_linear_form = 'HP ~ Attack * Defense * Speed * Legendary'
model4_linear_form += ' * Q("Sp. Def") * Q("Sp. Atk")'
# DO NOT try adding '* C(Generation) * C(Q("Type 1")) * C(Q("Type 2"))'
# That's 6*18*19 = 6*18*19 possible interaction combinations...
# ...a huge number that will blow up your computer

model4_spec = smf.ols(formula=model4_linear_form, data=pokeaman_train)
model4_fit = model4_spec.fit()
model4_fit.summary()
```

```python
yhat_model4 = model4_fit.predict(pokeaman_test)
y = pokeaman_test.HP
print("'In sample' R-squared:    ", model4_fit.rsquared)
print("'Out of sample' R-squared:", np.corrcoef(y,yhat_model4)[0,1]**2)
```

```python
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
import plotly.express as px
import plotly.graph_objects as go
from sklearn.model_selection import train_test_split

# Example dataset (pokeaman) assumed to be already loaded as 'pokeaman'

# Define linear form for model (adjusted for the context of 'pokeaman')
linear_form = 'HP ~ Attack + Defense + Speed + Legendary'

# Set number of repetitions for bootstrapping
reps = 100
in_sample_Rsquared = np.array([0.0] * reps)
out_of_sample_Rsquared = np.array([0.0] * reps)

# Loop for model fitting across multiple random splits
for i in range(reps):
    # Split the data into training and testing (50-50 split)
    pokeaman_train, pokeaman_test = train_test_split(pokeaman, train_size=0.5)
    
    # Fit the model on the training data
    model_fit = smf.ols(formula=linear_form, data=pokeaman_train).fit()
    
    # Calculate in-sample R-squared (on training data)
    in_sample_Rsquared[i] = model_fit.rsquared
    
    # Calculate out-of-sample R-squared (on testing data)
    y_pred = model_fit.predict(pokeaman_test)
    y_true = pokeaman_test['HP']
    out_of_sample_Rsquared[i] = np.corrcoef(y_true, y_pred)[0, 1]**2

# Create DataFrame for results
df = pd.DataFrame({
    "In Sample Performance (Rsquared)": in_sample_Rsquared,
    "Out of Sample Performance (Rsquared)": out_of_sample_Rsquared
})

# Plotting the results using Plotly
fig = px.scatter(df, 
                 x="In Sample Performance (Rsquared)", 
                 y="Out of Sample Performance (Rsquared)",
                 title="In-Sample vs Out-of-Sample Model Performance",
                 labels={"In Sample Performance (Rsquared)": "In-Sample R^2",
                         "Out of Sample Performance (Rsquared)": "Out-of-Sample R^2"})

# Add the line y=x for visual comparison
fig.add_trace(go.Scatter(x=[0, 1], y=[0, 1], name="y = x", line_shape='linear', line=dict(color='red', dash='dash')))

fig.show(renderer="png")
```

our goal is to see how the model performs on both the tested data and the additional general data, as we can see it is clustered around the y=x line which means that generally, the model is accurate in depicting a true HP predictor. we can see that sometimes the out-of-sample is more suited for the model than the in-sample and this could happen because sometimes the test data may be easier to predict than the training data, this means that the model might be very suited for the whole population which could lead to the out-of-sample having a higher r-squared value. it could be possible as well that the in-sample could capture too many outliers such that the the r-squared value might be lower for the training set.
